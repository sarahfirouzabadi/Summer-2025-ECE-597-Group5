```python
import pandas as pd
import numpy as np

data=pd.read_csv("/content/emails.csv")
data.info()
data.head(5)
```

```python
data.head(5)
```

```python
email_body=[]
for text in data['message']:
    parts=text.split("\n\n",1)
    if len(parts)>1:
        email_body.append(parts[1])
    else:
        email_body.append(text)
```

```python
print(data['message'][0])
```

```python
filtered_data = data[~data['message'].str.contains('enron.com', na=False)]

filtered_email_body = [body for body, original_message in zip(email_body, data['message']) if 'enron.com' not in original_message]

print("\nFiltered DataFrame head:")
print(filtered_data.head())

print("\nFirst few filtered email bodies:")
print(filtered_email_body[:5])
```

```python
print("\nNumber of rows in the filtered DataFrame:", len(filtered_data))
print("Number of items in the filtered email body list:", len(filtered_email_body))
```

```python
print("\nFirst 50 email bodies from the filtered DataFrame:")
for i in range(min(50, len(filtered_data))):
    print(f"\n--- Email Body {i+1} ---")
    print(filtered_data['message'].iloc[i])
```
